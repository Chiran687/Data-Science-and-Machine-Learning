<h1>Percentile</h1>

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sbs
```

```python
data = np.random.randint(90, size=(10))
data
```

```python
data = [36, 55, 56, 41, 37, 43,10,  6, 61, 50, 58]
```

```python
sorted_data = sorted(data)
```

```python
def compute_percentile(sorted_data, value):
    data_len_below_value = len(sorted_data[:sorted_data.index(value)])
    print(data_len_below_value)
    n = len(sorted_data)
    percentile_rank = data_len_below_value / n
    return percentile_rank
```

```python
percentile_rank_of_10 = compute_percentile(data, 6)
print(percentile_rank_of_10)
```
