## Hypothesis Testing
- A procedure for determining if an claim about a characteristic of a population is reasonable.
- In other words, `Hypothesis Testing` can be defined as formal decision-making procedure, which enables us to decide whether to accept or reject hypothesis.
- **What is Hypothesis?**
    - It is a statement or assumption made about a population parameter.
    - Two main types i.e.
        - `Null Hypothesis (HO)`
            - Deenoted by H0
            - Represents the default assumption / claim.
            - It is the hypothesis that you aim to test against.
            - Example: Average price of 1kg cauliflower in kathmandu is Rs 60 i.e. $H0: \mu = 60$
            
        - `Alternate Hypothesis (Ha or H1)`
            - Denoted by Ha or H1.
            - It is the inverse of Null Hypothesis.
            - Example: Average price of 1kg cauliflower in kathmandu is not equals to Rs 60 

- **Goal of Hypothesis Testing:**
    - To collect of sample data and use statistical methods to evaluate whether there is enough evidence to reject the null hypothesis in favour of Alternative Hypothesis.
    - Decision is based on test statistics, significance level that helps to decide whether to Reject Null Hypothesis or Fail to Reject Null Hypothesis. 
- **Example:**
    - Suppose someone says that the average price of the cauliflower/kg in kathmandu is Rs 60.
    - Is this statement True? How  would you decide? 
        - In reality you cannot take data from every retailors and wholesellers to validate your assumption.
    - What you will do is you will sample retailors around the kathmandu randomly and compare the average price to Rs 60.
    - The average price of sample retailors may not be exactly Rs 60 due to variability in price from one station to the next, also the sample size will be small compared to population size.
    - Suppose your average price was Rs 58. Is this  Rs 2 difference a result of chance variability or is the original assertion incorrect?
    - This can be answered via Hypothesis Test.

## What are different types of test
1. `z-test`
    - This is used for comparison of mean. 
    - Here population standard deviation is given.
    - We look into z-table to get the area under the curve for given level of significance.

2. `t-test`
    - This is also used for comparison of mean.
    - Here sample standard deviation of mean.
    - We look into t-table for given degree of freedom and level of significance.

3. `Annova Test`
    - In a annova test we do a analysis of Variance i.e. whether the variance is same or not.

4. `Chisquare Test`
    - Here we can compare categorical variables.
    - Useful for comparing several proportions.

5. `F-Test`
    - For testing hypothesis which involves more than two means, we use F-test.


## How to select the appropriate test statistics
- The choice among different test statistics depends on the type of the test we want to conduct, on the size of the sample and other information such as whether the population is normal, whether the population variance is known, etc

1. `If we know the variance of the population, we may apply the z-test irrespective of the normality of the population and irrespective of the sample size.`

2. `If the variance of the population is unknown, but the size of the sample is large (n>=30)`, 
    - We may still apply z-test since the estimate of the population variance from a large sample is a satisfactory estimate of the true population variance.

3. `If the variance of the population is unknown and our sample is small (n<30)`, 
    - We may apply t-test provided that the population is normal. 
    - For t-test normality is crucial.

4. `For independence and goodness of fit, we use chi-square test`. 
    - It is also used for comparing several proportions.

5. `For testing hypothesis which involves more than two means, we use F-test`.   
    - It is also used for testing overall significance of regression equation and ANOVA test. 
    - Normality of population should hold for the use of F-test.

## level of significance
- The level of significance ($\alpha$) is a critical statistical threshold used to determine when to reject the null hypothesis (HO).
- When a significance level of 5% ($\alpha$ = 0.05) is chosen, the null hypothesis is only rejected if the sample result is substantially different from the hypothesized value to a degree that it would occur by random chance with a probability of 0.05 or less.
- Note that if the 5% ($\alpha = 0.05$) is used, there is a probability of 0.05 of rejecting the null hypothesis when it is in fact true. This means, that there is 5% chance that you will accept your alternative hypothesis when your null hypothesis is actually true. 
    - 5% chance that the decision may be wrong.
- As discussed, If we reject a hypothesis when it should be accepted, we say that a `Type-I Error` has been made.
- The probability of `Type-I Error` is always equal to $\alpha$.
$$\alpha=P(H0 is rejected | H0 is True)$$

## Confidence Interval

## P Value
- It is the probability of obtaining a test statistic (or a more extreme test statistic) when you assume that the null hypothesis is True.

$$P(Obtaining-Test-Statistic | NULL-Hypothesis-is-True)$$

- `Conditions:`
    - If the p-value is less than or equal to the significance level ($\alpha$), we reject the null hypothesis.
    - If the p-value is greater than the significance level ($\alpha$), we fail to reject the null hypothesis.

- `Step2: State the alpha, i.e. alpha = 0.05`

- `Step3: State Decision Rule`
    - Find the range of confidence interval.
    - Full entire region of normal distribution is 1.
    - Region of Condfidence interval is: 1 - 0.05 => 0.95 (is the region of confidence interval)
    - Z-score:
        - it is used to find out how many standard deviation it is away from the mean.
    - In Z-test we will find out range using Z-table.

    - Z = x_mean - meu / sigma / sqrt(n)
    - Z_test = 14.60

`Step5: State the Results`  
    -  Z_test does not fall in between confidence interval, so reject the null hypothesis and we accept the alternate hypothesis.

## t-test 
- t-test is used in hypothesis testing, when population variance is unknown and sample size < 30.
- For t-test normality is crucial i.e. we may apply t-test provided that population data is normally distributed.
- `shape:`
    - t-distribution has bell shaped similar to the normal distribution with heavier tails.
    - exact shape of the t-distribution depends on degree of freedom.

- `Degree of Freedom (df):`
    - DF determines the shape of the t-distribution.
    - The larger the degree of freedom, the closer the t-distribution approximates the normal distribution.
    - Formulae: n-1

- `Small Samples:`
    - Useful for small samples (i.e. n < 30).
    - Increase in sample size, t-distribution approaches the normal distribution

- `Use Cases:`
    - Useful when population variance is unknown.
    - So population variance will be estimated from sample data.
    
- `test statistics`

  
- $$t_{\text{score}} = \frac{{\bar{X} - \mu}}{{\frac{{s}}{{\sqrt{n}}}}}$$
    
    When we sample from normal distribution, the obtained is t-distribution with n-1 degrees of freedom.

- `interpretation`
    - A positive t-score indicates that the sample mean is above the hypothesized mean.
    - A negative t-score indicates that the sample mean is below the hypothesized mean.



### Visualization (t-distribution)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t, norm

# Define degrees of freedom
# to visualize the t-distribution
degrees_of_freedom = [1, 2, 5, 10]

# Generate x values
x = np.linspace(-4, 4, 1000)

# Plot the t-distributions for different degrees of freedom
plt.figure(figsize=(10, 6))
for df in degrees_of_freedom:
    # Calculate the probability density function (PDF) for the t-distribution
    pdf = t.pdf(x, df)
    plt.plot(x, pdf, label=f'df = {df}')


# Plot the exact normal distribution
pdf_normal = norm.pdf(x)
plt.plot(x, pdf_normal, label='Normal Distribution', linestyle='--')

# Add labels and a legend
plt.title('t-Distribution')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

This shows the increase in degree of freedom, t-distributions is very close to normal distribution.

     $$t_{\text{score}} = \frac{{\bar{X} - \mu}}{{\frac{{s}}{{\sqrt{n}}}}}$$
     $$t_{\text{score}} = 11.96$$

- `Step 5: Decision`
    - $t_{\text{score}} > CI$
    - Reject Null Hypothesis

## Chi-Square Test

- Non-parametric test usually occurs with respect to population proportions.
- Whenever we have some kind of proportions of data at that point of time you cannot specificaly use a kind of parametric test, so you have to go with Non-parametric tst.

Q. In 2000 Indian Census, the age of the inidividual in a small town were found to be following: 

less than 18 --> 20%  
18 - 35 -------> 30%  
Greater than 35 --> 50%  

Considering this, In 2010 the group of sample, n = 500, individual were sampled.  

< 18 --------> 121  
18 - 35 -----> 288  
\> 35 --------> 91

Using $\alpha = 0.05$, Could you conclude population distribution of ages has changed in the last 10 yrs?

#### Step 1: Hypotheses

- Null Hypothesis (H0): The population distribution of ages in 2010 is the same as in 2000.
- Alternative Hypothesis (Ha): The population distribution of ages in 2010 is different from that in 2000.


#### Step 2: Set the Significance Level
- Significance Level ($\alpha$): 0.05

#### Step 3: Collect Data

In 2010, a sample of 500 individuals was collected, and their ages were categorized as follows:

- < 18: 121 individuals
- 18 - 35: 288 individuals
- \> 35: 91 individuals


#### Step 4: Calculate Expected Frequencies

We calculate the expected frequencies for each age group based on the 2000 census data:

- Expected frequency for "< 18" in 2010: $(E_1 = 500 \times 0.20 = 100)$
- Expected frequency for "18 - 35" in 2010: $(E_2 = 500 \times 0.30 = 150)$
- Expected frequency for "> 35" in 2010: $(E_3 = 500 \times 0.50 = 250)$


### Step 5: Calculate the Chi-Square Test Statistic

We calculate the chi-square test statistic for each category:

- For "< 18":
  $(\chi^2_1 = \frac{{(121 - 100)^2}}{100})$

- For "18 - 35":
  $(\chi^2_2 = \frac{{(288 - 150)^2}}{150})$

- For "> 35":
  $(\chi^2_3 = \frac{{(91 - 250)^2}}{250})$

Next, calculate the total chi-square statistic:  
$$(\chi^2_{\text{total}} = \chi^2_1 + \chi^2_2 + \chi^2_3 = 232.494)$$



#### Step 6: Determine Critical Value

We determine the critical chi-square value from a chi-square distribution with 2 degrees of freedom (df = 3 - 1 = 2) for  = 0.05.   
By looking to $\chi^2$ table, The critical $\chi^2$ value is approximately 5.991.


#### Step 7: Make a Decision

Since the calculated $\chi^2$ statistic (232.494) is greater than the critical chi-square value (5.991), we reject the null hypothesis.

#### Step 8: Conclusion

- Since the calculated $\chi^2$ statistic (217.05) is greater than the critical chi-square value (5.991), we reject the null hypothesis. 
- We conclude that the population distribution of ages in 2010 is different from that in 2000.   
- There is evidence to suggest that there has been a change in the distribution of ages in the last 10 years in the small town.
