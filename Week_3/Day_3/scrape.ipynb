```python
url = "https://www.worldometers.info/coronavirus/"
```

```python
import requests
response = requests.get(url)
print(response)
```

```python
print(response.content)
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(response.content, 'html.parser')
print(soup.prettify()[:5000])
```

```python
world_data = soup.find("tbody").find_all('tr')
```

```python
for data in world_data:
    print(data)
    break
```

```python
len(world_data)
```

```python
world_data[8]
```

```python
complete_data = []
for i in range(8, len(world_data)):
  data = []
  list_data = world_data[i].find_all("td")
  for i in list_data:
    data.append(i.text)
  complete_data.append(data)
```

```python
complete_data
```

```python
mapped_data = list(map(lambda x: x[1:10] + [x[12]] + [x[14]], complete_data))
```

```python
mapped_data
```

```python
sample_list = [1, 2]

square_lambda = lambda x: x**2
```

```python
print(square_lambda)
```

```python
squared_values = map(square_lambda, sample_list)
```

```python
for item in squared_values:
    print(item)
```

```python
column_names = [
    "Names",
    "Total Cases",
    "New Cases",
    "Total Deaths",
    "New Deaths",
    "Total Recovered",
    "New Recovered",
    "Active Cases",
    "Serious Cases",
    "Total tests",
    "Population"
]
```

```python
import pandas as pd

df = pd.DataFrame(mapped_data, columns = column_names)
df.to_csv('covid_data.csv', index=False)
```
