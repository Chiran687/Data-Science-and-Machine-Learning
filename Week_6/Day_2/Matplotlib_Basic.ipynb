<h2>1. How does the average satisfaction level change with increasing years of employment?</h2>

```python
import matplotlib.pyplot as plt
df.groupby('time_spend_company')['satisfaction_level'].mean().plot(kind='line')
plt.show()
```

<p><strong>The satisfaction_level was high initially but it dropped significiantly on fourth year and increased on upcomming year.</strong></p>

<h2>2. How many employees have spent a specific number of years (time spent in the company)</h2>

```python
ax= sns.barplot(df['time_spend_company'].value_counts().sort_index())
ax.bar_label(ax.containers[0], fontsize=10)
```

<p><li><strong>The max num of employees worked for 3 years.</strong></li>
<li><strong>The least num of employees woprked for 8 years.</strong></li>
</p>

<h2>3. What is the distribution of satisfaction levels among employees?</h2>

```python
df['satisfaction_level'].hist()
plt.xlabel('satisfaction_level')

plt.title('Distribution of satisfaction levels among employees.')
```

<h2>4. What is the distribution of work accidents among departments?</h2>

```python
plt.rcParams['text.color'] = '#000000'
plt.rcParams['axes.labelcolor']= '#909090'
plt.rcParams['xtick.color'] = '#909090'
plt.rcParams['ytick.color'] = '#909090'
plt.rcParams['font.size']=11
color_palette_list = ['#009ACD', '#ADD8E6', '#63D1F4', '#0EBFE9',   
                      '#C1F0F6', '#0099CC']

group = df.groupby('Department')['Work_accident'].sum()
plt.pie(group, labels=group.index, startangle=90, colors=color_palette_list, autopct='%1.1f%%',)
plt.title('Distribution of work accidents among departments')
plt.show()
```

<p><li><strong>Employees from Sales department have highest rate of accident.</li>
<li><strong>Employees from HR department have lowest rate of accident</li></p>

<h2>5 What is the distribution of promotions based on departments?</h2>

```python
# df.groupby('Department')['promotion_last_5years'].sum().plot(kind='bar', grid=True)
# plt.show()
sns.set_theme(rc={'figure.figsize':(11.7,8.27)})
ax= sns.barplot(df.groupby('Department')['promotion_last_5years'].sum())
ax.bar_label(ax.containers[0], fontsize=10)
```

<p><li><strong>Sales Department have highest rate of promotion.</li>
<li><strong>Product management department does not have promotion at all.</li></p>

<h2>6. How does employee satisfaction vary across different departments?`</h2>

```python
sns.set_theme(rc={'figure.figsize':(11.7,8.27)})
sns.boxplot(y=df['satisfaction_level'], x=df['Department'], width = 0.5)
```

<p><li><strong>Here employee satisfaction can be seen similar through all departments, having no outlier.</li></p>

<h2>8. How are the numerical variables correlated with each other?</h2>

```python
import seaborn as sns
new_df=pd.read_csv('../../Data/HR_comma_sep.csv', usecols=[0,1,2,3,4,5,6,7])

corr = new_df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Numerical Variables')
plt.show()
```

<h2>9. How to create different scatter plots under single matplotlib figure?</h2>

```python
import numpy as np
x = np.random.randint(50, size=(9))
y = np.random.randint(50, size=(9))

plt.subplot(2, 2, 1)
plt.scatter(x, y)

plt.subplot(2, 2, 2)
plt.scatter(x, y)

plt.show()
```
