### Single Linear Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
```

```python
df = pd.read_csv("../../Data/Salary_Data.csv")
df.head()
```

```python
sns.heatmap(df.corr())
```

```python
plt.scatter(df['Salary'], df['YearsExperience'])
plt.ylabel("YearsExperience")
plt.xlabel("Salary")
plt.show()
```

Converting to numpy array

```python
salary_np = df['Salary'].to_numpy()
years_np = df['YearsExperience'].to_numpy()
```

```python
salary_np.shape, years_np.shape
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression().fit(salary_np.reshape((30,1)), years_np)
predict = model.predict(salary_np.reshape((30,1)))
predict.shape
```

```python
plt.scatter(df['Salary'], df['YearsExperience'])
plt.plot(df['Salary'], predict, c='r')
plt.ylabel("YearsExperience")
plt.xlabel("Salary")
plt.show()
```

```python
predict_df = pd.DataFrame({'Years_Of_Experience': df['YearsExperience'],
                       'Salary': df['Salary'],
                       'Predicted_salary':predict})
predict_df.head()
```
