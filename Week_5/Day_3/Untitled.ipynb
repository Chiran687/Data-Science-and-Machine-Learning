iloc
* ways to slice pandas df via integer value
* it allows you to select data from a df using integer indices for both rows and columns
* 1st row is 0 and 2nd row is 1 and so on
* 1st column is 1, 2nd is 2 and so om

```python
df.iloc[0,0]
```

```python
df.iloc[:3, -3:]
```

loc
* loc is used for label-based indexing
* allows us to slice df using row label and column label instead of integer
* syntax df.loc[row_label, column_label]

```python
df.loc[0, 'satisfaction_level']
```

```python
df.loc[[0,1,2], ["satisfaction_level", "left", "salary"]]
```
